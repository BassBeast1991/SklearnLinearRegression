```python
# Load dataset using Pandas library and store in dataframe

import pandas as pd

cars = pd.read_csv('cars_dataset.csv')

# display first five rows
display(cars.head())
```

```python
# make a scatterplot for weight and horsepower columns

cars.plot.scatter(x='weight',y='horsepower')
```

```python
# make a scatterplot for weight and mpg columns

cars.plot.scatter(x='weight',y='mpg')
```

```python
# extract weight, horsepower and mpg

weight = cars[['weight']]
horsepower = cars['horsepower']

mpg = cars[['mpg']]
```

```python
# import LinearRegression and predict horsepower and mpg from weight

import sklearn

regr1 = sklearn.linear_model.LinearRegression()
regr1.fit(weight,horsepower)

regr2 = sklearn.linear_model.LinearRegression()
regr2.fit(weight,mpg)
```

```python
# print out slope coefficient and intercept of linear fits

print('Horsepower fit coefficient - ' + str(regr1.coef_))
print('Horsepower fit intercept - ' + str(regr1.intercept_))
print('\n')

print('MPG fit coefficient - ' + str(regr2.coef_))
print('MPG fit intercept - ' + str(regr2.intercept_))
print('\n')
```

```python
# make predictions for horsepower and mpg

horsepower_pred = regr1.predict(weight)
mpg_pred = regr2.predict(weight)
```

```python
# assign predictions to new columns in the dataframe

cars['predicted_horsepower'] = horsepower_pred
cars['predicted_mpg'] = mpg_pred
```

```python
# visualise the regression comparing actual and calculated values for horsepower

ax = cars.plot.scatter(x='weight', y='horsepower')
ax = cars.plot.line(x='weight', y='predicted_horsepower', ax=ax, c='red')

```

```python
# visualise the regression comparing actual and calculated values for mpg

ax = cars.plot.scatter(x='weight', y='mpg')
ax = cars.plot.line(x='weight', y='predicted_mpg', ax=ax, c='red')
```

```python
# assess goodness of fit using R2 score

R2_horsepower = regr1.score(weight, horsepower)
R2_mpg = regr2.score(weight, mpg)

print("The R2 score for horsepower is ", R2_horsepower)
print("The R2 score for MPG is ", R2_mpg)
```
